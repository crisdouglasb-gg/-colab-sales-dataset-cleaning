```python
# =========================
# CELDA 2: IMPORTAR LIBRERГЌAS
# =========================

import pandas as pd
import numpy as np

from langchain_core.documents import Document
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
# =========================
# CELDA 3: SUBIR ARCHIVO
# =========================

from google.colab import files

uploaded = files.upload()

df = pd.read_csv("sales_data_sample.csv", encoding="latin1")

df.head()
```

```python
# =========================
# CELDA 4: EXPLORACIГ“N DEL DATASET
# =========================

print("рџ“Љ Filas y columnas:", df.shape)

print("\nрџ“Њ Columnas:")
print(df.columns)

print("\nрџ“Њ Tipos de datos:")
print(df.dtypes)
```

```python
# =========================
# CELDA 5: DETECCIГ“N DE NaN
# =========================

print("\nрџ“Њ NaN por columna:")
print(df.isnull().sum())
```

```python
# =========================
# CELDA 6: ESTADГЌSTICAS
# =========================

print("\nрџ“Њ EstadГ­sticas:")
df.describe(include="all")
```

```python
# =========================
# CELDA 7: NORMALIZACIГ“N
# =========================

print("NaN ANTES:\n", df.isnull().sum())

num_cols = df.select_dtypes(include=[np.number]).columns
cat_cols = df.select_dtypes(exclude=[np.number]).columns

# numГ©ricos в†’ media
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

# categГіricos в†’ UNKNOWN
df[cat_cols] = df[cat_cols].fillna("UNKNOWN")
```

```python
# =========================
# CELDA 8: VALIDACIГ“N
# =========================

print("\nNaN DESPUГ‰S:\n", df.isnull().sum())

print("Total NaN finales:", df.isnull().sum().sum())
```

```python
# =========================
# CELDA 9: LIMPIEZA DE COLUMNAS
# =========================

df.columns = df.columns.str.lower().str.replace(" ", "_")

df.head()
```

```python
# =========================
# CELDA 10: EXPORTAR CSV
# =========================

df.to_csv("sales_data_clean.csv", index=False)

print("Dataset limpio guardado")
```

```python
# =========================
# CELDA 11: RESUMEN FINAL
# =========================

print("Shape final:", df.shape)
print("NaN final total:", df.isnull().sum().sum())
```
