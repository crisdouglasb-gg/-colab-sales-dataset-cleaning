```python
df = pd.read_csv("sales_data_sample.csv", encoding="latin1")

df.head()
```

```python
print("📊 Filas y columnas:", df.shape)

print("\n📌 Columnas:")
print(df.columns)

print("\n📌 Tipos de datos:")
print(df.dtypes)

print("\n📌 NaN por columna:")
print(df.isnull().sum())

print("\n📌 Estadísticas:")
df.describe(include="all")
```

```python
#NORMALIZACION
print("NaN ANTES:\n", df.isnull().sum())

num_cols = df.select_dtypes(include=[np.number]).columns
cat_cols = df.select_dtypes(exclude=[np.number]).columns

# numéricos → media
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

# categóricos → UNKNOWN
df[cat_cols] = df[cat_cols].fillna("UNKNOWN")

print("\nNaN DESPUÉS:\n", df.isnull().sum())
```

```python
# VALIDACION
print("Total NaN finales:", df.isnull().sum().sum())
```

```python
# LIMPIEZA DE COLUMNAS
df.columns = df.columns.str.lower().str.replace(" ", "_")

df.head()
```

```python
# GUARDAR DATASET LIMPIO
df.to_csv("sales_data_clean.csv", index=False)

print("Dataset limpio guardado")
```
