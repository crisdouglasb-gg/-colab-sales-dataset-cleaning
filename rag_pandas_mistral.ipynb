```python
# =========================
# LEER CSV
# =========================

df = pd.read_csv("sales_data_sample.csv", encoding="latin1")

print("Dataset cargado correctamente")

df.head()
```

```python
# =========================
# NORMALIZACIÓN
# =========================

# eliminar duplicados
df = df.drop_duplicates()

# rellenar valores nulos
df = df.fillna("DESCONOCIDO")

# limpiar nombres de columnas
df.columns = df.columns.str.lower()

df.columns = df.columns.str.replace(" ", "_")

print("Datos normalizados")

df.head()
```

```python
# =========================
# IMPORTACIONES RAG
# =========================

from langchain_core.documents import Document

from langchain_community.vectorstores import FAISS

from langchain_community.embeddings import HuggingFaceEmbeddings

print("Importaciones RAG correctas")
```

```python
# =========================
# CONVERTIR CSV A DOCUMENTOS
# =========================

documentos = []

for index, row in df.iterrows():

    texto = " | ".join(row.astype(str))

    documentos.append(
        Document(page_content=texto)
    )

print("Cantidad de documentos creados:")
print(len(documentos))
```

```python
# =========================
# MODELO DE EMBEDDINGS
# =========================

embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

print("Modelo de embeddings cargado correctamente")
```

```python
# =========================
# INSTALAR FAISS
# =========================

!pip install -q faiss-cpu
```

```python
# =========================
# CREAR BASE VECTORIAL
# =========================

vectorstore = FAISS.from_documents(
    documentos,
    embedding_model
)

print("Base vectorial FAISS creada correctamente")
```

```python
# =========================
# CREAR RETRIEVER
# =========================

retriever = vectorstore.as_retriever(
    search_kwargs={"k": 5}
)

print("Retriever configurado correctamente")
```

```python
# =========================
# PRUEBA RETRIEVER
# =========================

documentos_recuperados = retriever.invoke(
    "producto más vendido"
)

print(documentos_recuperados[0].page_content)
```

```python
# =========================
# CONSULTA RAG
# =========================

consulta = "¿Cuál es el producto más vendido?"

contexto = retriever.invoke(consulta)

texto_contexto = "\n".join(
    [doc.page_content for doc in contexto]
)

prompt = f"""
Usá el siguiente contexto para responder la pregunta.

Contexto:
{texto_contexto}

Pregunta:
{consulta}
"""

respuesta = llm.invoke(prompt)

print(respuesta.content)
```
